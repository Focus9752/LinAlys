```python
import sympy as sp
import matplotlib.pyplot as plt

# Plot f(x,y)=sin(x)sin(y)
x, y = sp.symbols('x y')
f = sp.sin(x)*sp.sin(y)
p = sp.plotting.plot3d(f, (x, -1, 1), (y, -1, 1), show=False)

# Save to pdf
p.save(r'C:\Users\marcu\OneDrive\Documents\GitHub\LinAlys\UA10\opg1b plot.pdf')

# Draw a contour plot around (0,pi)

```
