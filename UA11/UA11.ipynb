```python
# Plot mængden x^2 + 2y^2 <= 1, x >= 0
# med 1000 punkter i x-retningen og 1000 punkter i y-retningen
# og farvelæg de punkter, der opfylder betingelsen.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)
Z = X**2 + 2*Y**2

plt.figure(1)

plt.contourf(X, Y, Z, levels=[0, 1], colors=['blue', 'red'])

# Vis x = 0 og y = 0
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# Få plottet til at se utroligt smukt ud :D
plt.grid()
plt.tight_layout()
plt.axis('equal')

plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot af $x^2 + 2y^2 \leq 1$, $x \geq 0$")

plt.show()
```
