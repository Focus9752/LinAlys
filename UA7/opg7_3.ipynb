Importer nødvendige funktioner/symboler

```python
import sympy as sp


from sympy.abc import v, theta
from sympy.plotting import plot_implicit
```

Definer E ud fra variablerne

```python
E = 1/2 * v ** 2 - 10 * sp.cos(theta)
E
```

Vi ønsker at lave niveaukurver, vi følger metoden beskrevet [her.](https://python-intro.nbi.ku.dk/notebooks/sympy/Notebook_FlereVar_plot.html#niveaukurver):

```python
# Her kan du angive en liste af værdier af E, som hver giver en niveaukurve
værdier_for_E = [1,5,15,20,25,35]

# Sæt størrelsen af billedet
v_range       = (-10, 10)
theta_range   = (-10, 10)

# Vi laver en liste til figurer
figurer   = []

# Vi sætter et loop op over vores værdier for c
for i in range(len(værdier_for_E)):
    
    # Lav et nyt implicit plot for hver værdi af c:
    ny_figur = plot_implicit(sp.Eq(E, værdier_for_E[i]),              # Opskriv ligheden for E = E_i 
                             (v, v_range[0], v_range[1]),             # Sæt grænserne for variable
                             (theta, theta_range[0], theta_range[1]),
                             line_color = f"C{i}",                    # Vælge den i'te farve i matplotlib
                             show       = False)                      # Ikke vis
    

    # Tilføj denne figur til en liste
    figurer.append(ny_figur)
```

Vi har nu 5 figurer, vi kan samle dem ved at loope over og tilføje dem til det første:

```python
# Her kan du angive en liste af værdier af E, som hver giver en niveaukurve
værdier_for_E = [1,5,15,25,35,10]

# E_O lader til at være 10

# Sæt størrelsen af billedet
v_range       = (-10, 10)
theta_range   = (-10, 10)
# Vi laver en liste til figurer
figurer   = []

# Vi sætter et loop op over vores værdier for c
for i in range(len(værdier_for_E)):
    
    # Lav et nyt implicit plot for hver værdi af c:
    ny_figur = plot_implicit(sp.Eq(E, værdier_for_E[i]),              # Opskriv ligheden for E = E_i 
                             (v, v_range[0], v_range[1]),             # Sæt grænserne for variable
                             (theta, theta_range[0], theta_range[1]),
                             line_color = f"C{i}",                    # Vælge den i'te farve i matplotlib
                             show       = False)                      # Ikke vis
    

    # Tilføj denne figur til en liste
    figurer.append(ny_figur)

vis_figur = figurer[0]    #Lav en ny figur ud fra den første figur i listen

# Loop nu over resten af listen ved at indeksere med [1:]
for tilføj_figur in figurer[1:]:
    vis_figur.append(tilføj_figur[0])          # Tilføj til den første figur
    
# Vis figuren
vis_figur.show()
```
