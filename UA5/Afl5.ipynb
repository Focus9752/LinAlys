```python
import sympy as sp
from sympy import *
from mpmath import mp

# Opgave 5.1
# b)
x = symbols("x")
T_5 = sp.series(sp.sin(x**2), x, 0, 11)
T_10 = sp.series(sp.sin(x**2), x, 0, 21)

T_5 = T_5.removeO()
T_10 = T_10.removeO()

plot = sp.plot(sp.sin(x**2), T_5, T_10, ylim=[-2,2])
plot.legend = True
plot[0].label = r"$sin(x^2)$"
plot[1].label = "5. grads Taylorpolynomie"
plot[2].label = "10. grads Taylorpolynomie"

plot.save(r"C:\Users\marcu\OneDrive\Documents\GitHub\LinAlys\UA5\plots.pdf")
```
