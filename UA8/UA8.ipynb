```python
# Let z = 1/2 + 3/4 i
# Write a program to plot z^(-8), z^(-7), ..., z^7, z^8 in the complex plane
import math
import numpy as np
import matplotlib.pyplot as plt


# z = 1/2 + 3/4 i
z = complex(0.5, 0.75)
# z^(-8), z^(-7), ..., z^7, z^8
z_list = [z**i for i in range(-8, 9)]

# Create a new figure and a new axes instance
fig, ax = plt.subplots(figsize=(6,6))

# Plot the complex numbers with colors based on their order
scatter = ax.scatter([z.real for z in z_list], [z.imag for z in z_list], c=range(len(z_list)), cmap='viridis')

# Draw the real and imaginary axes in the middle
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)

# Set the axes
ax.axis('equal')
ax.set_xlim(-2.5, 2.5)
ax.set_ylim(-2.5, 2.5)

# Add grid lines
ax.grid(True, which='both', color='grey', linewidth=0.25)

# Add labels
ax.set_xlabel(r'$\mathbb{R}$')
ax.set_ylabel(r'$\mathrm{i}$')
ax.set_title('Plot af z^n for n = -8, -7, ..., 7, 8', pad=15)
ax.text(0.5, 1.02, 'Farven på punkterne angiver rækkefølgen de optræder i.', transform=ax.transAxes, fontsize=8, ha='center')

# Add a colorbar
fig.colorbar(scatter, ax=ax, orientation='vertical')

plt.show()

# Save to pdf
fig.savefig('UA8.pdf', bbox_inches='tight')
```
