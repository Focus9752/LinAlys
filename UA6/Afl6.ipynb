```python
import sympy as sp
from sympy import *
from mpmath import mp

# Opgave 6.3
# b)
v = symbols("v")

m = 1
c = 300

expr = (m*c**2)/(sp.sqrt(1-(v**2)/(c**2)))

T_2 = sp.series(expr, v, 0, 3).removeO()
T_4 = sp.series(expr, v, 0, 5).removeO()
T_10 = sp.series(expr, v, 0, 11).removeO()

plot = sp.plot(expr, T_2, T_4, T_10, (v, 0, 300), ylim=(0, 400000), xlabel="v", ylabel="E(v)")

plot.legend = True
plot[0].label = r"E"
plot[1].label = r"$T_2E$"
plot[2].label = r"$T_4E$"
plot[3].label = r"$T_{10}E$"

plot.save(r"C:\Users\marcu\OneDrive\Documents\GitHub\LinAlys\UA6\plots.pdf")



```
