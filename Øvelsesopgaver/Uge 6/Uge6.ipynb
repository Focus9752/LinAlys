```python
import numpy as np
import matplotlib.pyplot as plt

# TK 2.15 (S 104)
# a)
f = lambda x, y : 3*x*y + y**2

x, y = np.arange(-5,5), np.arange(-5,5)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.figure()
CS = plt.contourf(X,Y,Z)
plt.plot(1,2, f(1,2), 2)
plt.show()
```
